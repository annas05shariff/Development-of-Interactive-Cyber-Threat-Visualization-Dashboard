# Object-Oriented Programming (OOP)

Object-Oriented Programming (OOP) is a programming paradigm that organizes code using objects and classes.  
It helps in structuring programs in a modular and reusable way.

---

## 1. Class
A class is a blueprint or template used to create objects.  
It defines the attributes (variables) and methods (functions) that the objects will have.

---

## 2. Object
An object is an instance of a class created using the class blueprint.  
It contains actual values for the attributes and can use the class methods.

---

## 3. Encapsulation
Encapsulation is the process of bundling data and methods inside a single unit (class).  
It helps in protecting data from unauthorized access and improves security.

---

## 4. Inheritance
Inheritance allows one class to acquire properties and methods from another class.  
It promotes code reusability and reduces redundancy in programs.

---

## 5. Polymorphism
Polymorphism allows the same method name to behave differently for different objects.  
It improves flexibility and allows one interface to be used for multiple functionalities.

---

## Application in This Project
In this cyber threat dashboard project, OOP is used to model cyber attacks as objects.  
Each attack contains attributes like IP address, attack type, and severity level.


```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
attack_types = ["Brute Force", "SQL Injection", "Phishing", "Ransomware", "DDoS"]

random_attacks = [random.choice(attack_types) for _ in range(10)]

print("Random Attack List:")
print(random_attacks)
```

```python
attack_tuple = tuple(random_attacks)

print("Converted to Tuple:")
print(attack_tuple)
```

```python
unique_attacks = set(random_attacks)

print("Unique Attack Types:")
print(unique_attacks)
```

```python
unique_attacks = set(random_attacks)

print("Unique Attack Types:")
print(unique_attacks)
```

```python
attack_dict = {}

for i in range(5):
    ip = f"192.168.1.{i}"
    attack = random.choice(attack_types)
    attack_dict[ip] = attack

print("Dictionary (IP -> Attack Type):")
print(attack_dict)
```

```python
class CyberAttack:
    def __init__(self, ip, attack_type, severity):
        self.ip = ip
        self.attack_type = attack_type
        self.severity = severity

    def display(self):
        print(f"IP: {self.ip}")
        print(f"Attack Type: {self.attack_type}")
        print(f"Severity: {self.severity}")

# Create object with random severity
attack1 = CyberAttack("192.168.1.100", random.choice(attack_types), random.randint(1,5))
attack1.display()
```

```python
severities = np.random.randint(1, 6, size=10)

print("Random Severity Levels:")
print(severities)
```

```python
data = {
    "Attack_Type": random_attacks,
    "Severity": severities
}

df = pd.DataFrame(data)

print("Cyber Threat DataFrame:")
print(df)
```

```python
attack_counts = df["Attack_Type"].value_counts()

plt.bar(attack_counts.index, attack_counts.values)
plt.title("Cyber Attack Distribution")
plt.xlabel("Attack Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```
